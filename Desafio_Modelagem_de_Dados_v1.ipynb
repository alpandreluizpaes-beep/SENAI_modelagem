###1.3. Descrição dos atributos

**ESTADO**
*   id_estado: número que representa a unidade federativa, tipo int
*   nome: nome completo da unidade federativa, tipo varchar
*   sigla: sigla da unidade federativa, tipo char(2)
*   regiao: nome região do Brasil onde a unidade federativa se encontra, tipo varchar

**SEXO**
*   id_sexo: número que representa o gênero biológico do feto, tipo int
*   descricao: descricao (masculino ou feminino), tipo varchar

**CAUSA**
*   id_causa: código alfanumérico que representa a causa do óbito do feto, tipo char(2)
*   descricao: descrição detalhada da causa do óbito do feto, tipo varchar

**ESTABELECIMENTO**
*   id_local: número que representa o local declarado do óbito do feto, tipo int
*   descricao: local de declaração do óbito do feto, tipo varchar

**OBITOS**
*   id_obito: número que identifica o registro do óbito, tipo int
*   peso_feto: massa do feto em gramas, tipo int
*   data: data da declaração de óbito do feto, tipo date
*   idade_mae: idade em anos da mãe no momento do óbito do feto, tipo int
*   semana_gestacao: semana de gestação no momento do óbito do feto, tipo int

```python
## Conexão

import mysql.connector as mysql
from mysql.connector import Error
import pandas as pd
import psycopg2
import sys
import time
import requests
from io import BytesIO

myconn = mysql.connect(host='localhost', user='root', password='!Qwerty1234')
mycur = myconn.cursor()

## Função def para consultas futuras

def consulta_db(sql):
    mycur.execute(sql)
    return mycur.fetchall()


## Criação do banco
mycur.execute("CREATE DATABASE IF NOT EXISTS SENAI")
mycur.execute("USE SENAI")
```

```python
## Criação das tabelas

comandos = [

"""
CREATE TABLE IF NOT EXISTS ESTADO (
    id_estado INT PRIMARY KEY,
    nome VARCHAR(19) NOT NULL,
    sigla CHAR(2) NOT NULL,
    regiao VARCHAR(12) NOT NULL,
    UNIQUE (nome, sigla)
)
""",

"""
CREATE TABLE IF NOT EXISTS SEXO (
    id_sexo INT PRIMARY KEY,
    descricao VARCHAR(9) NOT NULL,
    UNIQUE (descricao)
)
""",

"""
CREATE TABLE IF NOT EXISTS CAUSA (
    id_causa CHAR(3) PRIMARY KEY,
    descricao VARCHAR(44) NOT NULL,
    UNIQUE (descricao)
)
""",

"""
CREATE TABLE IF NOT EXISTS ESTABELECIMENTO (
    id_local INT PRIMARY KEY,
    descricao VARCHAR(19) NOT NULL,
    UNIQUE (descricao)
)
""",

"""
CREATE TABLE IF NOT EXISTS OBITOS (
    id_obito INT PRIMARY KEY AUTO_INCREMENT,
    idade_mae INT,
    semana_gestacao INT,
    peso_feto INT,
    data DATE,
    id_estado INT,
    id_sexo INT,
    id_causa CHAR(3),
    id_local INT,
    FOREIGN KEY (id_estado) REFERENCES ESTADO(id_estado),
    FOREIGN KEY (id_sexo) REFERENCES SEXO(id_sexo),
    FOREIGN KEY (id_causa) REFERENCES CAUSA(id_causa),
    FOREIGN KEY (id_local) REFERENCES ESTABELECIMENTO(id_local)
)
"""
]

for comando in comandos:
    mycur.execute(comando)

myconn.commit()

print("Banco SENAI e tabelas criadas com sucesso!")
print("Tabelas existentes:")
print(consulta_db("SHOW TABLES;"))
```

###2.4. População do banco de dados

```python
mycur.execute("USE SENAI")
```

```python
## POPULAR TABELA ESTADO

df_estado = df[['ocor_CODIGO_UF',
                'ocor_NOME_UF',
                'ocor_SIGLA_UF',
                'ocor_REGIAO']].drop_duplicates()

df_estado = df_estado.dropna()

dados_estado = df_estado.values.tolist()

sql_insert_estado = """
INSERT IGNORE INTO ESTADO
(id_estado, nome, sigla, regiao)
VALUES (%s, %s, %s, %s)
"""

mycur.executemany(sql_insert_estado, dados_estado)
myconn.commit()

print("Tabela ESTADO populada!")
print("Total inseridos:", mycur.rowcount)
```

```python
## POPULAR TABELA SEXO

df_sexo = df[['SEXO', 'def_sexo']].drop_duplicates()
df_sexo = df_sexo.dropna()

dados_sexo = df_sexo.values.tolist()

sql_insert_sexo = """
INSERT IGNORE INTO SEXO
(id_sexo, descricao)
VALUES (%s, %s)
"""

mycur.executemany(sql_insert_sexo, dados_sexo)
myconn.commit()

print("Tabela SEXO populada!")
print("Total inseridos:", mycur.rowcount)
```

```python
## POPULAR TABELA CAUSA

df_causa = df[['id_causa', 'CAUSA']].drop_duplicates()
df_causa = df_causa.dropna()

dados_causa = df_causa.values.tolist()

sql_insert_causa = """
INSERT IGNORE INTO CAUSA
(id_causa, descricao)
VALUES (%s, %s)
"""

mycur.executemany(sql_insert_causa, dados_causa)
myconn.commit()

print("Tabela CAUSA populada!")
print("Total inseridos:", mycur.rowcount)
```

```python
## POPULAR TABELA ESTABELECIMENTO

df_estab = df[['LOCOCOR', 'def_loc_ocor']].drop_duplicates()
df_estab = df_estab.dropna()

dados_estab = df_estab.values.tolist()

sql_insert_estab = """
INSERT IGNORE INTO ESTABELECIMENTO
(id_local, descricao)
VALUES (%s, %s)
"""

mycur.executemany(sql_insert_estab, dados_estab)
myconn.commit()

print("Tabela ESTABELECIMENTO populada!")
print("Total inseridos:", mycur.rowcount)
```

```python
## POPULAR TABELA OBITOS

from datetime import datetime
import pandas as pd
import numpy as np

## Converter DTOBITO (30012019 → 2019-01-30)
df['DTOBITO'] = (
    df['DTOBITO']
    .astype(str)
    .str.zfill(8)
)

df['DTOBITO'] = pd.to_datetime(
    df['DTOBITO'],
    format='%d%m%Y',
    errors='coerce'  # se tiver erro vira NaT
)

## Substituir NaT por None (MySQL entende como NULL)
df['DTOBITO'] = df['DTOBITO'].where(df['DTOBITO'].notnull(), None)

## Substituir NaN por None no DataFrame inteiro
df = df.replace({np.nan: None})

## Preparar INSERT

sql_insert = """
INSERT INTO OBITOS
(idade_mae, semana_gestacao, peso_feto, data,
 id_estado, id_sexo, id_causa, id_local)
VALUES (%s, %s, %s, %s, %s, %s, %s, %s)
"""

dados_para_inserir = []

for _, row in df.iterrows():
    dados_para_inserir.append((
        row['IDADEMAE'],
        row['SEMAGESTAC'],
        row['PESO'],
        row['DTOBITO'],
        row['ocor_CODIGO_UF'],
        row['SEXO'],
        row['id_causa'],
        row['LOCOCOR']
    ))

## Inserção em lote

mycur.executemany(sql_insert, dados_para_inserir)
myconn.commit()

print("Tabela OBITOS populada!")
print("Total inseridos:", mycur.rowcount)
```
